```python
# 라이브러리 사용
import tensorflow as tf
import pandas as pd
```

```python
# 데이터 준비
파일경로 = 'https://raw.githubusercontent.com/blackdew/tensorflow1/master/csv/lemonade.csv'
데이터 = pd.read_csv(파일경로)
데이터.head()
```

```python
# 종속변수, 독립변수
독립 = 데이터[['온도']]
종속 = 데이터[['판매량']]

print(독립.shape, 종속.shape)
```

```python
# 모델은 만듭니다.
x = tf.keras.layers.Input(shape=[1]) # 컬럼이 1개이기에 shape 1
y = tf.keras.layers.Dense(1)(x)
model = tf.keras.models.Model(x, y)
model.compile(loss='mse')
```

```python
# 모델을 학습합니다.
model.fit(독립, 종속, epochs=10000, verbose=0) # verbose = 0, 출력제외
```

```python
# 모델을 이용합니다.
model.predict(독립)
```
